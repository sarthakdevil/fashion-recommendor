```python
from PIL import Image
from transformers import TextStreamer
FastVisionModel.for_inference(model)  # Make sure this is called

# Load the image (already uploaded)
image_path = "image.jpg"
image = Image.open(image_path).convert("RGB")  # Ensure it's in correct format

instruction = '''You are a fashion expert. Suggest complementary outfit items for the item shown in the image. Use short keywords. Start with "Suggestions:" and format as:
- Item: <Name>
- Colors: <Color Suggestions>
- Suggestions:
  - <Item 1> - <Color>
  - <Item 2> - <Color>
  '''

# Create multimodal messages properly
messages = [
    {
        "role": "user",
        "content": [
            {"type": "image", "image": image},
            {"type": "text", "text": instruction}
        ]
    }
]

# Apply chat template
input_text = tokenizer.apply_chat_template(messages, add_generation_prompt=True)

# Tokenize + preprocess
inputs = tokenizer(
    image,
    input_text,
    add_special_tokens=False,
    return_tensors="pt",
).to("cuda")

# Stream the output
text_streamer = TextStreamer(tokenizer, skip_prompt=True)
_ = model.generate(
    **inputs,
    streamer=text_streamer,
    max_new_tokens=128,
    use_cache=True,
    temperature=1.5,
    top_p=0.9,
)
```

```python
from PIL import Image

instruction = '''You are a fashion expert. Suggest complementary outfit items for the item shown in the image. Use short keywords. Start with "Suggestions:" and format as:
- Item: <Name>
- Colors: <Color Suggestions>
- Suggestions:
  - <Item 1> - <Color>
  - <Item 2> - <Color>
'''

def convert_to_conversation(sample):
    if not isinstance(sample, dict):
        raise TypeError(f"Expected 'sample' to be a dictionary, but got {type(sample).__name__}")

    image_path = sample.get("image_path", "").lstrip("/")
    recommendation = sample.get("recommendation", "No recommendation available.")

    try:

        image = Image.open(image_path).convert("RGB")
    except Exception as e:
        print(f"Warning: Could not load image at '{image_path}'. Error: {e}")
        image = None  # Optionally, provide a default image

    conversation = [
        {
            "role": "user",
            "content": [
                {"type": "text", "text": instruction},
                {"type": "image", "image": image} if image else {"type": "text", "text": "Image not available."}
            ]
        },
        {
            "role": "assistant",
            "content": [
                {"type": "text", "text": recommendation}
            ]
        }
    ]

    return {"messages": conversation}
```
